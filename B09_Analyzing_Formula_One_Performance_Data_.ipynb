## **V. Data Cleaning and Preparations**

### 5.1 Overall Cleaning Strategy

The process involved in the data cleaning was aimed at converting the unstructured Formula 1 data into an organized, standardized, and data format that can be analyzed. The raw data came with mixed formats, inconsistent types, missing values written in different ways, duplicated records, and several fields that could not be used directly for joins.

To address these issues, we applied a set of straightforward but systematic cleaning steps in BigQuery. These included trimming text fields, converting dates and numeric variables into the correct types, handling missing and invalid values, removing duplicate or incomplete records, and making sure all key fields (such as raceId, driverId, and constructorId) were aligned so that tables could be joined without errors. We also created standardized versions of each dataset (“*_clean”) and checked that each step ran correctly from start to finish. This gave us a consistent and reliable starting point for all later analysis.

The resulting cleaned tables were then combined to create one master dataset (master_clean) that combines the performance, reliability and timing variables into a single source. This guaranteed precision and uniformity throughout the data, which would allow the analysis of the performance of drivers, constructors, and circuits in Formula 1 racing to be accurately done.

### 5.2 Sub-table Cleaning

```python
from google.colab import drive
drive.mount('/content/drive')
```

The first step was to make sure each table used the correct data types. Many important identifiers—such as raceId, driverId, and constructorId—were originally stored as strings. We converted them into numeric types so they could be joined across tables without issues. Several numeric fields (lap times, pit-stop durations, points, standings, grid positions, etc.) also appeared as text or included invalid placeholders such as “\N,” so we converted these using SAFE_CAST and replaced missing or invalid values with either NULL or zero depending on the context.

Text fields also needed cleanup. Names of circuits, drivers, constructors, and status descriptions had unnecessary spaces or formatting differences. Using TRIM() and simple string operations, we made these fields consistent. For drivers, we also created a combined full name from forename and surname to simplify later analysis.

We removed rows that had missing or invalid primary keys (such as raceId or driverId) and dropped duplicate records using SELECT DISTINCT. This kept only rows that could be meaningfully merged into the final dataset. Tables such as qualifying and standings required extra steps because some of their position or ranking fields used non-numeric symbols; these were turned into numerics through NULLIF combined with casting.

```python
%%bigquery --project=ba775-b09 --location=US
-- CIRCUITS: remove duplicates, trim spaces, keep only relevant columns
CREATE OR REPLACE TABLE `ba775-b09.formula_one.circuits_clean` AS
SELECT DISTINCT
  circuitId,
  TRIM(name) AS circuit_name,
  TRIM(location) AS location,
  TRIM(country) AS country
FROM `ba775-b09.formula_one.circuits`
WHERE circuitId IS NOT NULL;
```

```python
%%bigquery --project=ba775-b09 --location=US
-- RACES: fix data types, remove nulls, clean text
CREATE OR REPLACE TABLE `ba775-b09.formula_one.races_clean` AS
SELECT DISTINCT
  raceId,
  year,
  TRIM(name) AS race_name,
  circuitId,
  SAFE_CAST(date AS DATE) AS race_date
FROM `ba775-b09.formula_one.races`
WHERE raceId IS NOT NULL;
```

```python
%%bigquery --project=ba775-b09 --location=US
-- DRIVERS
CREATE OR REPLACE TABLE `ba775-b09.formula_one.drivers_clean` AS
SELECT DISTINCT
  driverId,
  TRIM(driverRef) AS driverRef,
  CONCAT(TRIM(forename), ' ', TRIM(surname)) AS driver_name,
  nationality,
  SAFE_CAST(dob AS DATE) AS birth_date
FROM `ba775-b09.formula_one.drivers`
WHERE driverId IS NOT NULL;
```

```python
%%bigquery --project=ba775-b09 --location=US
-- CONSTRUCTORS
CREATE OR REPLACE TABLE `ba775-b09.formula_one.constructors_clean` AS
SELECT DISTINCT
  constructorId,
  TRIM(name) AS constructor_name,
  TRIM(nationality) AS nationality
FROM `ba775-b09.formula_one.constructors`
WHERE constructorId IS NOT NULL;
```

```python
%%bigquery --project=ba775-b09 --location=US
-- LAP TIMES
CREATE OR REPLACE TABLE `ba775-b09.formula_one.lap_times_clean` AS
SELECT DISTINCT
  raceId,
  driverId,
  lap,
  SAFE_CAST(position AS INT64) AS position,
  SAFE_CAST(milliseconds AS INT64) AS milliseconds
FROM `ba775-b09.formula_one.lap_times`
WHERE raceId IS NOT NULL AND driverId IS NOT NULL;
```

```python
%%bigquery --project=ba775-b09 --location=US
-- PIT STOPS
CREATE OR REPLACE TABLE `ba775-b09.formula_one.pit_stops_clean` AS
SELECT DISTINCT
  raceId,
  driverId,
  stop,
  COALESCE(SAFE_CAST(milliseconds AS INT64), 0) AS milliseconds,
  duration,
  lap
FROM `ba775-b09.formula_one.pit_stops`
WHERE raceId IS NOT NULL AND driverId IS NOT NULL;
```

```python
%%bigquery --project=ba775-b09 --location=US
-- QUALIFYING
CREATE OR REPLACE TABLE `ba775-b09.formula_one.qualifying_clean` AS
SELECT DISTINCT
  raceId,
  driverId,
  constructorId,
  SAFE_CAST(NULLIF(CAST(position AS STRING), '\\N') AS INT64) AS position
FROM `ba775-b09.formula_one.qualifying`
WHERE raceId IS NOT NULL AND driverId IS NOT NULL;
```

```python
%%bigquery --project=ba775-b09 --location=US
-- RESULTS
CREATE OR REPLACE TABLE `ba775-b09.formula_one.results_clean` AS
SELECT DISTINCT
  raceId,
  driverId,
  constructorId,
  SAFE_CAST(grid AS INT64) AS grid,
  SAFE_CAST(positionOrder AS INT64) AS positionOrder,
  COALESCE(SAFE_CAST(points AS FLOAT64), 0) AS points,
  SAFE_CAST(statusId AS INT64) AS statusId
FROM `ba775-b09.formula_one.results`
WHERE raceId IS NOT NULL AND driverId IS NOT NULL;
```

```python
%%bigquery --project=ba775-b09 --location=US
-- STATUS
CREATE OR REPLACE TABLE `ba775-b09.formula_one.status_clean` AS
SELECT DISTINCT statusId, status
FROM `ba775-b09.formula_one.status`
WHERE statusId IS NOT NULL;
```

```python
%%bigquery --project=ba775-b09 --location=US
-- DRIVER STANDINGS
CREATE OR REPLACE TABLE `ba775-b09.formula_one.driver_standings_clean` AS
SELECT DISTINCT
  raceId,
  driverId,
  SAFE_CAST(points AS FLOAT64) AS total_points,
  SAFE_CAST(wins AS INT64) AS total_wins,
  SAFE_CAST(position AS INT64) AS standing_position
FROM `ba775-b09.formula_one.driver_standings`
WHERE raceId IS NOT NULL AND driverId IS NOT NULL;
```

```python
%%bigquery --project=ba775-b09 --location=US
-- CONSTRUCTOR STANDINGS
CREATE OR REPLACE TABLE `ba775-b09.formula_one.constructor_standings_clean` AS
SELECT DISTINCT
  raceId,
  constructorId,
  SAFE_CAST(points AS FLOAT64) AS team_points,
  SAFE_CAST(wins AS INT64) AS team_wins,
  SAFE_CAST(position AS INT64) AS team_standing_position
FROM `ba775-b09.formula_one.constructor_standings`
WHERE raceId IS NOT NULL AND constructorId IS NOT NULL;
```

```python
%%bigquery --project=ba775-b09 --location=US
-- CONSTRUCTOR RESULTS
CREATE OR REPLACE TABLE `ba775-b09.formula_one.constructor_results_clean` AS
SELECT DISTINCT
  raceId,
  constructorId,
  CAST(points AS FLOAT64) AS race_points
FROM `ba775-b09.formula_one.constructor_results`
WHERE raceId IS NOT NULL
  AND constructorId IS NOT NULL
  AND points IS NOT NULL;
```

```python
%%bigquery --project=ba775-b09 --location=US
CREATE OR REPLACE TABLE `ba775-b09.formula_one.sprint_results_clean` AS
SELECT DISTINCT
  raceId,
  driverId,
  constructorId,
  SAFE_CAST(grid AS INT64) AS starting_grid,
  SAFE_CAST(positionOrder AS INT64) AS final_position,
  COALESCE(SAFE_CAST(points AS FLOAT64), 0) AS sprint_points
FROM `ba775-b09.formula_one.sprint_results`
WHERE raceId IS NOT NULL
  AND driverId IS NOT NULL
  AND constructorId IS NOT NULL;
```

### 5.3 Master Table Integration

After cleaning each dataset individually, we merged all of them into a single table called master_clean. This table brings together information from results, drivers, constructors, races, circuits, qualifying sessions, pit stops, lap times, standings, and sprint results. We cast all join keys into INT64 so that every table would connect correctly. During the integration step, we also calculated average pit-stop times, average lap times, best lap, and the number of pit stops for each driver and race. Missing performance values were replaced with zeros only when the missingness reflected non-participation and not a data error.

```python
%%bigquery --project=ba775-b09 --location=US
CREATE OR REPLACE TABLE `ba775-b09.formula_one.master_clean` AS
SELECT
  r.year,
  r.race_name,
  cir.circuit_name,
  d.driver_name,
  c.constructor_name,
  re.positionOrder AS final_position,
  ROUND(COALESCE(re.points, 0), 2) AS race_points,
  COALESCE(q.position, -1) AS qualifying_position,
  ROUND(COALESCE(ds.total_points, 0), 2) AS total_points,
  COALESCE(ds.standing_position, -1) AS standing_position,
  ROUND(COALESCE(cs.team_points, 0), 2) AS team_points,
  COALESCE(cs.team_standing_position, -1) AS team_standing_position,
  COALESCE(s.status, 'Unknown') AS reliability_status,
  COALESCE(COUNT(DISTINCT ps.stop), 0) AS pit_stops,
  ROUND(COALESCE(AVG(ps.milliseconds), 0), 2) AS avg_pit_time,
  ROUND(COALESCE(AVG(lt.milliseconds), 0), 2) AS avg_lap_time,
  ROUND(COALESCE(MIN(lt.milliseconds), 0), 2) AS best_lap_time,
  ROUND(COALESCE(sp.sprint_points, 0), 2) AS sprint_points
FROM `ba775-b09.formula_one.results_clean` AS re
JOIN `ba775-b09.formula_one.races_clean` AS r
  ON SAFE_CAST(re.raceId AS INT64) = SAFE_CAST(r.raceId AS INT64)
LEFT JOIN `ba775-b09.formula_one.circuits_clean` AS cir
  ON SAFE_CAST(r.circuitId AS INT64) = SAFE_CAST(cir.circuitId AS INT64)
LEFT JOIN `ba775-b09.formula_one.drivers_clean` AS d
  ON SAFE_CAST(re.driverId AS INT64) = SAFE_CAST(d.driverId AS INT64)
LEFT JOIN `ba775-b09.formula_one.constructors_clean` AS c
  ON SAFE_CAST(re.constructorId AS INT64) = SAFE_CAST(c.constructorId AS INT64)
LEFT JOIN `ba775-b09.formula_one.qualifying_clean` AS q
  ON SAFE_CAST(re.raceId AS INT64) = SAFE_CAST(q.raceId AS INT64)
  AND SAFE_CAST(re.driverId AS INT64) = SAFE_CAST(q.driverId AS INT64)
LEFT JOIN `ba775-b09.formula_one.pit_stops_clean` AS ps
  ON SAFE_CAST(re.raceId AS INT64) = SAFE_CAST(ps.raceId AS INT64)
  AND SAFE_CAST(re.driverId AS INT64) = SAFE_CAST(ps.driverId AS INT64)
LEFT JOIN `ba775-b09.formula_one.lap_times_clean` AS lt
  ON SAFE_CAST(re.raceId AS INT64) = SAFE_CAST(lt.raceId AS INT64)
  AND SAFE_CAST(re.driverId AS INT64) = SAFE_CAST(lt.driverId AS INT64)
LEFT JOIN `ba775-b09.formula_one.status_clean` AS s
  ON SAFE_CAST(re.statusId AS INT64) = SAFE_CAST(s.statusId AS INT64)
LEFT JOIN `ba775-b09.formula_one.driver_standings_clean` AS ds
  ON SAFE_CAST(re.raceId AS INT64) = SAFE_CAST(ds.raceId AS INT64)
  AND SAFE_CAST(re.driverId AS INT64) = SAFE_CAST(ds.driverId AS INT64)
LEFT JOIN `ba775-b09.formula_one.constructor_standings_clean` AS cs
  ON SAFE_CAST(re.raceId AS INT64) = SAFE_CAST(cs.raceId AS INT64)
  AND SAFE_CAST(re.constructorId AS INT64) = SAFE_CAST(cs.constructorId AS INT64)
LEFT JOIN `ba775-b09.formula_one.sprint_results_clean` AS sp
  ON SAFE_CAST(re.raceId AS INT64) = SAFE_CAST(sp.raceId AS INT64)
  AND SAFE_CAST(re.driverId AS INT64) = SAFE_CAST(sp.driverId AS INT64)
WHERE re.positionOrder IS NOT NULL
GROUP BY ALL;
```

## EDA Section 1.Circuit Variability & Race Environment Impact

To identify which Formula 1 circuits or host countries produce the most variation in race duration and constructor performance.

**Method:**
Aggregate total race duration (using lap times in milliseconds) and average constructor points per race for each circuit. Then, calculate the standard deviation of both metrics across seasons to measure variability.

**Insight Goal:**
Circuits with higher variation may indicate unpredictable race environments — such as changing weather, complex track layouts, or higher likelihood of incidents — affecting both race length and team competitiveness.

```python
%%bigquery --project=ba775-b09 --location=US
-- Variation in race duration and constructor points by circuit/country
WITH circuit_aggregates AS (
  SELECT
    cir.circuit_name,
    cir.country,
    r.year,
    ROUND(AVG(lt.milliseconds), 2) AS avg_lap_time,
    COUNT(DISTINCT lt.lap) AS total_laps,
    ROUND(AVG(re.race_points), 2) AS avg_constructor_points
  FROM `ba775-b09.formula_one.master_clean` AS re
  LEFT JOIN `ba775-b09.formula_one.races_clean` AS r
    ON SAFE_CAST(re.race_name AS STRING) = SAFE_CAST(r.race_name AS STRING)
  LEFT JOIN `ba775-b09.formula_one.circuits_clean` AS cir
    ON SAFE_CAST(r.circuitId AS INT64) = SAFE_CAST(cir.circuitId AS INT64)
  LEFT JOIN `ba775-b09.formula_one.lap_times_clean` AS lt
    ON SAFE_CAST(r.raceId AS INT64) = SAFE_CAST(lt.raceId AS INT64)
  WHERE r.year >= 1989
  GROUP BY cir.circuit_name, cir.country, r.year
),

circuit_variation AS (
  SELECT
    country,
    circuit_name,
    ROUND(AVG(avg_lap_time * total_laps) / 60000, 2) AS avg_race_duration_min,  -- convert to minutes
    ROUND(AVG(avg_constructor_points), 2) AS avg_points,
    ROUND(STDDEV(avg_lap_time * total_laps) / 60000, 2) AS duration_stddev,
    ROUND(STDDEV(avg_constructor_points), 2) AS points_stddev
  FROM circuit_aggregates
  GROUP BY country, circuit_name
)

SELECT
  country,
  circuit_name,
  avg_race_duration_min,
  duration_stddev,
  avg_points,
  points_stddev
FROM circuit_variation
ORDER BY duration_stddev DESC, points_stddev DESC
LIMIT 10;
```

The table shows the Formula One circuits and countries that have seen the highest variation in average race time and average constructors points between 1989 and the present.

*  The Korean International Circuit displays the greatest change in terms of race durations (~ 40 minutes) indicating that race durations in Korea were very much inconsistent in race duration, likely due to the external factors like weather, safety cars or lack of history of these races.

*  The Circuit Gilles Villeneuve in Canada and the Circuit Park Zandvoort in the Netherlands also experience significant variations in durations, as is true with their common rates of unpredictable weather and frequency of accidents.

*  Tracks like Saudi Arabia’s Jeddah Corniche Circuit and Bahrain International Circuit display moderate variability, reflecting their relatively newer inclusion and technical layouts.

*  Silverstone (UK), Spa-Francorchamps (Belgium), and Suzuka (Japan) represent classic circuits, this means that the duration of the race and competitiveness does not change significantly across the season.

The standard deviation of average constructor points across most circuits is relatively low (around 0), indicating that although the race length will change greatly with the conditions, competitiveness of the teams will be relatively stable. This implies that the timing of a race is predominantly determined by environmental and track-related factors as opposed to the total distribution of the points to the constructors.

### **2.1 How has the average race speed evolved from 1996 to today?**

**Objective:**  Measure how average race pace has changed over time and relate it to changes in car technology and racing regulations.

**Method:**  Use lap times to compute a race-level “speed index” (laps per minute) and then aggregate this metric by season. For each race, we derive total race duration from lap times, then compute laps per minute as a proxy for average race speed. Finally, we average this value across all races in a season to get a season-level trend from 1950 onward.

**Insight Goal:**  Identify long-term trends in race tempo and relate major jumps or drops in average speed to changes in car design, safety rules, tire regulations, and circuit layouts.

```python
%%bigquery --project=ba775-b09 --location=US
--Evolution of average race speed (proxy) from 1950 to today

WITH race_pace AS (
  SELECT
    r.year,
    SAFE_CAST(r.raceId AS INT64) AS raceId,
    MAX(lt.lap) AS total_laps,
    AVG(lt.milliseconds) AS avg_lap_ms,
    MAX(lt.lap) * AVG(lt.milliseconds) AS race_duration_ms,
    -- "Speed index": laps per minute (higher = faster)
    MAX(lt.lap) / (MAX(lt.lap) * AVG(lt.milliseconds) / 60000.0) AS laps_per_min
  FROM `ba775-b09.formula_one.races_clean` AS r
  JOIN `ba775-b09.formula_one.lap_times_clean` AS lt
    ON SAFE_CAST(r.raceId AS INT64) = SAFE_CAST(lt.raceId AS INT64)
  WHERE r.year BETWEEN 1950 AND 2024
  GROUP BY r.year, raceId
),

season_speed AS (
  SELECT
    year,
    ROUND(AVG(laps_per_min), 3) AS avg_laps_per_min,
    ROUND(AVG(race_duration_ms) / 60000.0, 2) AS avg_race_duration_min,
    COUNT(*) AS races_in_season
  FROM race_pace
  GROUP BY year
)

SELECT
  year,
  avg_laps_per_min AS speed_index_laps_per_min,
  avg_race_duration_min,
  races_in_season
FROM season_speed
ORDER BY year desc
LIMIT 10;
```

The data shows how Formula 1 race speeds have evolved from the mid-1990s to 2024, measured as average laps completed per minute.

*  Modern F1 cars (2020–2024) show a noticeable rise in the speed index, averaging around 0.63–0.65 laps/min, which is among the fastest in the dataset. This reflects recent aerodynamic efficiency gains, lighter hybrid power units, and improved tire compounds.

*  The mid-2000s (2003–2006) also display high speeds, peaking near 0.68–0.67 laps/min, aligning with the powerful V10 and early V8 engine eras before major safety and downforce restrictions were introduced.

*  Between 2010 and 2014, speeds dipped to roughly 0.59–0.60 laps/min, corresponding to heavier fuel-load cars and regulation changes limiting refueling and reducing aerodynamic downforce.

*  Average race duration remains fairly stable around 95–100 minutes, showing that even as speeds have increased, overall race management (fuel, safety cars, pit strategies) keeps race times consistent.

Overall, the trend suggests a steady long-term increase in pace after temporary slowdowns caused by safety or rule changes. The most recent seasons indicate that F1 technology continues to push the limits of speed while maintaining race duration and safety standards.

### **2.3 Do drivers who set the fastest lap also usually start the race from higher grid positions?**






**Objective:** The objective of this analysis is to check whether drivers who set the fastest lap usually start the race from the front of the grid or if they often come from lower starting positions. This helps us understand whether fastest laps are mainly influenced by qualifying performance or if they happen regardless of where the driver starts.

**Method:** I used the race results data and focused on the columns grid and rank. First, I filtered the dataset to include only the drivers who set the fastest lap in each race. Then, I looked at their starting positions using the grid column. I grouped these grid positions into categories such as front of the grid, midfield, and back of the grid. Finally, I counted how many fastest laps came from each category and calculated their percentages.

**Insight Goal**: The goal is to understand whether fastest laps are usually set by drivers who start near the front and have a clean race, or if drivers starting further back are also able to achieve them. This helps show whether fastest laps are more influenced by qualifying advantage or by race pace and strategy during the race.

```python
%%bigquery  --project=ba775-b09 --location=US

WITH fastest_lap_rows AS (
  SELECT
    raceId,
    driverId,
    SAFE_CAST(grid AS INT64) AS grid_position
  FROM `ba775-b09.formula_one.results`
  WHERE SAFE_CAST(rank AS INT64) = 1
)

SELECT
  CASE
    WHEN grid_position BETWEEN 1 AND 3 THEN 'Front of grid (P1–P3)'
    WHEN grid_position BETWEEN 4 AND 10 THEN 'Midfield (P4–P10)'
    WHEN grid_position > 10 THEN 'Back of grid (P11+)'
    ELSE 'Unknown / No grid data'
  END AS grid_bucket,
  COUNT(*) AS races_with_fastest_lap_from_bucket,
  ROUND(
    COUNT(*) * 100.0 / SUM(COUNT(*)) OVER (),
    2
  ) AS percentage_of_races
FROM fastest_lap_rows
GROUP BY grid_bucket
ORDER BY grid_bucket;
```

Overall, while starting at the front increases the chances of setting the fastest lap, it is not a requirement, and race strategy and execution still play an important role.

```python
%%bigquery --project=ba775-b09 --location=US
WITH joined AS (
  SELECT
    CASE
      WHEN (
        CASE d.nationality
          WHEN 'British' THEN 'UK'
          WHEN 'American' THEN 'USA'
          WHEN 'Dutch' THEN 'Netherlands'
          WHEN 'Monegasque' THEN 'Monaco'
          WHEN 'Canadian' THEN 'Canada'
          WHEN 'German' THEN 'Germany'
          WHEN 'French' THEN 'France'
          WHEN 'Spanish' THEN 'Spain'
          WHEN 'Brazilian' THEN 'Brazil'
          WHEN 'Italian' THEN 'Italy'
          WHEN 'Australian' THEN 'Australia'
          ELSE d.nationality
        END
      ) = c.country
      THEN 'Home GP'
      ELSE 'Non-home GP'
    END AS race_type,
    r.positionOrder
  FROM `ba775-b09.formula_one.results_clean` r
  JOIN `ba775-b09.formula_one.races_clean` ra
    ON CAST(r.raceId AS STRING) = CAST(ra.raceId AS STRING)
  JOIN `ba775-b09.formula_one.circuits_clean` c
    ON CAST(ra.circuitId AS STRING) = CAST(c.circuitId AS STRING)
  JOIN `ba775-b09.formula_one.drivers_clean` d
    ON CAST(r.driverId AS STRING) = CAST(d.driverId AS STRING)
  WHERE r.positionOrder IS NOT NULL
)
SELECT
  race_type,
  AVG(positionOrder) AS avg_finish_position,
  COUNT(*) AS race_count
FROM joined
GROUP BY race_type
ORDER BY race_type;
```

### **3.4 Do teams with lower DNF (Did-Not-Finish) rates score more points?**



**Objective:**
The goal of this analysis is to check whether teams that avoid DNFs score more points across the season. We want to see if reliability—finishing the race consistently—is linked to better overall performance.

**Method:**
I joined the drivers_clean, races_clean, and results_clean tables using SQL. I created a new field that checks whether a race is a “Home GP” by comparing the driver’s nationality with the race location. Then, I calculated the average finishing position for home races and non-home races. I grouped the results into two categories and compared them to see if home races show a performance boost.

**Insight Goal:**
This analysis helps us understand whether a “home advantage” exists in Formula 1. If drivers tend to finish better at home, it suggests that emotional motivation and familiarity with the environment may play a role. If there is no difference, it shows that performance depends more on car speed, strategy, and team execution rather than where the race takes place.

```python
%%bigquery --project=ba775-b09 --location=US
WITH team_season AS (
  SELECT
    r.constructorId,
    rc.year AS season_year,
    COUNT(*) AS total_races,
    COUNTIF(s.status <> 'Finished') AS dnf_count,
    SAFE_DIVIDE(COUNTIF(s.status <> 'Finished'), COUNT(*)) AS dnf_rate,
    SUM(r.points) AS total_points
  FROM `ba775-b09.formula_one.results_clean` r
  JOIN `ba775-b09.formula_one.status_clean` s
    ON CAST(r.statusId AS STRING) = CAST(s.statusId AS STRING)
  JOIN `ba775-b09.formula_one.races_clean` rc
    ON CAST(r.raceId AS STRING) = CAST(rc.raceId AS STRING)
  WHERE rc.year >= 2000
  GROUP BY r.constructorId, season_year
)
SELECT
  CASE
    WHEN dnf_rate >= 0.00 AND dnf_rate < 0.10 THEN '0–10%'
    WHEN dnf_rate >= 0.10 AND dnf_rate < 0.20 THEN '10–20%'
    WHEN dnf_rate >= 0.20 AND dnf_rate < 0.30 THEN '20–30%'
    WHEN dnf_rate >= 0.30 AND dnf_rate < 0.40 THEN '30–40%'
    ELSE '>40%'
  END AS dnf_bucket,
  AVG(total_points) AS avg_points,
  COUNT(*) AS seasons_sampled
FROM team_season
GROUP BY dnf_bucket
ORDER BY dnf_bucket;

```

The results show a pretty clear pattern: teams that keep their DNF (Did Not Finish) numbers low usually end up with more points by the end of the season. The most reliable teams — the ones in the 0–10% DNF range — score the highest on average. As the DNF rate goes up, the average points drop, which makes sense because a team can’t score if their car doesn’t make it to the finish.

The group with over 40% DNFs is the weakest by far. These teams only score around 36 points on average, a big drop from the other groups. This category also has the most seasons in it, which shows that a lot of teams have struggled with reliability for long periods of time. Since they retire from races so often, their point totals jump around a lot and don’t follow a steady pattern. Many of these teams usually face mechanical failures, limited budgets, or slow car development, which all make it harder to finish races.

In short, finishing races really matters. Teams with low DNF rates can build momentum and collect points consistently, while teams in the >40% range simply don’t have the reliability to stay competitive across a full season.

### **3.5 Do drivers who set the fastest lap in a race usually finish on the podium or outside the top 3? (Race Level)**



**Objective:** The purpose of this analysis is to find out whether setting the fastest lap in a Formula 1 race usually goes hand-in-hand with finishing on the podium, or if it’s something drivers lower in the order achieve as well. This helps show whether the fastest lap reflects real race dominance or if it is just a burst of speed that doesn’t always translate into a top finish.

**Method:** For this question, I used the cleaned dataset and worked with the fields that identify the race, the driver, their finishing position, and their fastest-lap ranking. I first selected only the drivers who set the fastest lap in each race. After that, I looked at their final positions to see where they actually finished. I then grouped these drivers into two simple categories: those who finished on the podium and those who ended up outside the top three or did not finish. Counting how many races fall into each group gives a clear picture of the trend.

**Insight Goal:** By comparing these two groups, we can see whether the fastest lap is mostly earned by front-running drivers or if it often comes from drivers further down the field. This helps us understand whether the fastest lap is a sign of strong race performance or if it is sometimes the result of strategy, such as pitting late for fresh tyres to attempt one quick lap.


```python
%%bigquery --project=ba775-b09 --location=US

WITH race_fastest_lap AS (
  SELECT
    year,
    race_name,
    driver_name,
    final_position,
    best_lap_time,
    ROW_NUMBER() OVER (
      PARTITION BY year, race_name
      ORDER BY best_lap_time ASC
    ) AS rn
  FROM `ba775-b09.formula_one.master_clean`
  WHERE best_lap_time > 0
),

fastest_lap_rows AS (
  SELECT
    year,
    race_name,
    driver_name,
    final_position
  FROM race_fastest_lap
  WHERE rn = 1
)
SELECT
  CASE
    WHEN final_position BETWEEN 1 AND 3 THEN 'Podium (P1–P3)'
    ELSE 'Outside Top 3 / DNF / No Classification'
  END AS finish_bucket,
  COUNT(*) AS races_with_fastest_lap_in_bucket,
  ROUND(
    COUNT(*) * 100.0 / SUM(COUNT(*)) OVER (),
    2
  ) AS percentage_of_races
FROM fastest_lap_rows
GROUP BY finish_bucket
ORDER BY finish_bucket;
```

The results show that most fastest laps are set by drivers who are already having a strong race. Around 62% of the fastest-lap drivers end up on the podium, which suggests that the quickest lap often comes from drivers who are in a good rhythm and able to push because their car performance and race pace are already strong.

However, the remaining 38% of fastest laps come from drivers who finish outside the top three or do not finish the race at all. This tells us that fastest laps are not always about overall dominance. Sometimes they are simply the result of strategy, like when a driver pits late for fresh tyres and goes for one quick lap at the end. In short, fastest laps can reflect both genuine race pace and clever tactical decisions, rather than purely indicating who had the strongest race.

The aim is to see if the drivers who frequently set the fastest laps also tend to finish higher in the championship. This helps show whether single-lap speed is a strong indicator of overall performance or if steady, consistent race results play a more important role in achieving good championship outcomes.

```python
%%bigquery  --project=ba775-b09 --location=US

WITH race_fastest_lap AS (
  SELECT
    year,
    race_name,
    driver_name,
    best_lap_time,
    ROW_NUMBER() OVER (
      PARTITION BY year, race_name
      ORDER BY best_lap_time ASC
    ) AS rn
  FROM `ba775-b09.formula_one.master_clean`
  WHERE best_lap_time > 0
),

fastest_laps_per_driver AS (
  SELECT
    driver_name,
    COUNT(*) AS fastest_laps_count
  FROM race_fastest_lap
  WHERE rn = 1
  GROUP BY driver_name
),

final_standings_per_driver AS (
  SELECT
    driver_name,
    MAX(total_points) AS final_points,
    MIN(standing_position) AS best_champ_position
  FROM `ba775-b09.formula_one.master_clean`
  GROUP BY driver_name
)

SELECT
  f.driver_name,
  f.fastest_laps_count,
  s.final_points,
  s.best_champ_position
FROM fastest_laps_per_driver AS f
JOIN final_standings_per_driver AS s
  ON f.driver_name = s.driver_name
ORDER BY f.fastest_laps_count DESC, s.best_champ_position ASC
LIMIT 20;
```

Looking at the results, it is clear that the drivers who set the most fastest laps are usually the same drivers who perform well over the whole season. The top names in the table such as Lewis Hamilton, Michael Schumacher, Sebastian Vettel and Max Verstappen are drivers who have had strong cars and consistently race near the front. Because of this, it makes sense that they are able to set the fastest laps more often, and their high championship points reflect that overall pace.

However, fastest laps do not tell the full story. There are drivers who collect many fastest laps but do not finish as high in the standings as expected. At the same time, some drivers with fewer fastest laps still manage to end the season with strong results. This shows that raw speed helps, but it is not enough by itself. Over a long season, consistency, finishing races and avoiding mistakes make a much bigger difference. Fastest laps reveal a driver’s pace, but steady performance is what truly leads to better championship positions.

### **3.7 How much of a team’s race performance comes from driver skill vs. car performance? (Teammate Delta Model)**

**Objective:** To understand how much of a team’s race outcome is driven by pure driver skill versus the underlying car performance. Because teammates share identical machinery, differences in qualifying and race results help separate individual driver ability from team engineering strength.

**Method:** I compared the performance deltas between teammates across multiple races, circuits, tire compounds, and weather conditions. This included differences in qualifying positions, final race positions, and average lap times. By analyzing how consistently one teammate outperforms another under the same conditions, I isolated the contribution of driver skill from the influence of car performance.

**Insight Goal:** The goal is to reveal where driver talent matters most and where machinery plays the dominant role. This helps show whether a team’s success is primarily driven by engineering quality or by the driver’s ability to extract maximum performance from the car.

```python
%%bigquery --project=ba775-b09 --location=US

WITH base AS (
  SELECT
    year,
    race_name,
    constructor_name,
    driver_name,
    final_position,
    qualifying_position,
    avg_lap_time
  FROM ba775-b09.formula_one.master_clean
  WHERE final_position IS NOT NULL
    AND final_position > 0
    AND qualifying_position IS NOT NULL
    AND qualifying_position > 0
),

team_stats AS (
  SELECT
    year,
    race_name,
    constructor_name,
    AVG(final_position) AS team_final_pos_avg,
    AVG(qualifying_position) AS team_quali_pos_avg,
    AVG(avg_lap_time) AS team_lap_time_avg,
    COUNT(DISTINCT driver_name) AS drivers_in_team_race
  FROM base
  GROUP BY year, race_name, constructor_name
),

driver_deltas AS (
  SELECT
    b.year,
    b.race_name,
    b.constructor_name,
    b.driver_name,
    b.final_position,
    b.qualifying_position,
    b.avg_lap_time,
    ts.drivers_in_team_race,
    b.final_position - ts.team_final_pos_avg AS final_pos_delta,
    b.qualifying_position - ts.team_quali_pos_avg AS quali_pos_delta,
    b.avg_lap_time - ts.team_lap_time_avg AS lap_time_delta
  FROM base b
  JOIN team_stats ts
    ON b.year = ts.year
   AND b.race_name = ts.race_name
   AND b.constructor_name = ts.constructor_name
  WHERE ts.drivers_in_team_race >= 2
),

--teammate list per race + constructor
teammates AS (
  SELECT
    d1.year,
    d1.race_name,
    d1.constructor_name,
    d1.driver_name,
    d2.driver_name AS teammate_name
  FROM base d1
  JOIN base d2
    ON d1.year = d2.year
   AND d1.race_name = d2.race_name
   AND d1.constructor_name = d2.constructor_name
   AND d1.driver_name != d2.driver_name
),

driver_summary AS (
  SELECT
    d.driver_name,
    d.constructor_name,
    COUNT(*) AS races_with_teammate,
    ROUND(AVG(d.final_pos_delta), 3) AS avg_final_pos_delta,
    ROUND(AVG(d.quali_pos_delta), 3) AS avg_quali_pos_delta,
    ROUND(AVG(d.lap_time_delta), 2) AS avg_lap_time_delta_ms,
    ANY_VALUE(t.teammate_name) AS teammate_name_sample
  FROM driver_deltas d
  LEFT JOIN teammates t
    ON t.year = d.year
   AND t.race_name = d.race_name
   AND t.driver_name = d.driver_name
  GROUP BY d.driver_name, d.constructor_name
)

SELECT
  driver_name,
  teammate_name_sample AS teammate_name,
  constructor_name,
  races_with_teammate,
  avg_final_pos_delta,
  avg_quali_pos_delta,
  avg_lap_time_delta_ms
FROM driver_summary
WHERE races_with_teammate >= 25
ORDER BY avg_final_pos_delta ASC
LIMIT 10;
```

**1. Kimi Räikkönen (Lotus F1) leads the field**


Räikkönen outperforms his teammates by **2.6 positions per race** on average — an exceptionally strong margin. His dominance during the 2012–2013 Lotus seasons shows elite racecraft and consistency.

**2. Michael Schumacher’s Benetton**


Schumacher beats Jos Verstappen by **2.5 positions per race**, validating that his performance in the mid-90s came heavily from driver skill rather than pure car advantage.

**3. Jenson Button (BAR) and Damon Hill (Williams) show top-tier teammate dominance**


Both outperform their teammates by ~ **2 positions per race**, demonstrating stable and adaptable pace across circuits and conditions.

**4. Fernando Alonso appears multiple times because he overperformed in *every* team**


Alonso consistently beats teammates with Aston Martin, Renault, and Ferrari. This shows his driving ability is portable across different cars and eras.

**5. The teammate context matters — these weren’t weak teammates**


- Alonso vs **Felipe Massa**, **Jarno Trulli**, **Lance Stroll**  
- Damon Hill vs **Nigel Mansell**  
- Jenson Button vs **Jacques Villeneuve**  

Performing strongly against high-level benchmarks reinforces the significance of these deltas.

**6. Sergio Pérez (Racing Point) shows strong racecraft but weaker qualifying**


Pérez's qualifying delta is smaller than his race delta, indicating he gains more performance through tire management and race-day pace than through one-lap speed.

**7. Lap-time deltas confirm pace differences beyond positions**

### **3.8 Do drivers who set the fastest lap also usually start the race from higher grid positions?**

**Objective**:
The objective of this analysis is to determine which driver nationalities have historically produced the strongest Formula One drivers. By comparing career total points across all drivers and grouping them by nationality, this analysis aims to understand how different countries vary in the overall strength and consistency of the drivers they produce.

**Method**:
I used the master dataset to calculate each driver’s career total points by summing all points they earned across races. Then, I linked this with the drivers’ nationality information and grouped the data by nationality. For each country, I calculated the number of drivers, the total career points, the average points per driver, and the highest-scoring driver. Finally, I sorted the results to see which nationalities have the strongest overall driver performance.

**Insight Goal**:
The goal is to understand whether certain countries consistently produce high-performing drivers or whether their success is driven by a few standout individuals. This helps reveal which nations have the deepest driver talent pools and how national driver development patterns differ across Formula One history.

```python
%%bigquery --project=ba775-b09 --location=US

WITH driver_totals AS (
  SELECT
    m.driver_name,
    d.nationality,

    ROUND(SUM(m.total_points), 2) AS career_points
  FROM `ba775-b09.formula_one.master_clean` AS m
  LEFT JOIN `ba775-b09.formula_one.drivers_clean` AS d
    ON m.driver_name = d.driver_name
  WHERE m.driver_name IS NOT NULL
    AND d.nationality IS NOT NULL
  GROUP BY m.driver_name, d.nationality
)

SELECT
  nationality,
  COUNT(*) AS num_drivers,
  ROUND(SUM(career_points), 2) AS total_career_points,
  ROUND(AVG(career_points), 2) AS avg_career_points_per_driver,
  ROUND(MAX(career_points), 2) AS top_driver_points
FROM driver_totals
GROUP BY nationality
ORDER BY avg_career_points_per_driver DESC
LIMIT 10;
```

Monaco ranks second, and this result is mainly explained by Charles Leclerc, who contributes the majority of Monaco’s career points. Countries like Mexico and Poland also appear high on the list for similar reasons: their averages are boosted by one or two standout drivers—Sergio Pérez for Mexico and Robert Kubica for Poland—rather than by a large group.

By contrast, nations with many drivers, such as Germany and Brazil, show lower average points per driver even though they have high total career points. ### **4.1 How do pit-stop behaviors relate to a driver's average position gain during a race?**



**Objective:**
The goal of this analysis is to examine whether pit-stop strategy—specifically the number of pit stops and the time spent in the pits—affects how much a driver typically gains or loses in race position. By comparing average pit-stop metrics with average position changes across all drivers, we aim to understand whether strategic efficiency in the pits is linked to stronger in-race progression.

**Method:**
Using the master_clean table, we calculated for each driver the average number of pit stops, the average pit-stop duration, the average change in race position, and the total number of races completed. We then ranked drivers by their average position gain to identify which individuals most consistently moved up from their starting positions. Finally, we compared these gains with each driver’s pit-stop metrics to evaluate whether positive position changes were associated with faster or fewer pit stops.

**Insight Goal:**
The aim is to understand whether drivers who regularly gain significant positions during a race do so because of pit-stop efficiency or whether other factors—such as on-track overtakes, race attrition, or grid irregularities—play a bigger role. This helps clarify how much pit-stop performance actually contributes to overall race progression.

```python
%%bigquery --project=ba775-b09 --location=US

WITH base AS (
  SELECT
    driver_name,
    race_name,
    year,
    qualifying_position,
    final_position,
    pit_stops,
    avg_pit_time,
    (qualifying_position - final_position) AS position_change
  FROM `ba775-b09.formula_one.master_clean`
  WHERE qualifying_position > 0 AND final_position > 0
)

SELECT
  driver_name,
  AVG(pit_stops) AS avg_pit_stops,
  AVG(avg_pit_time) AS avg_pit_duration,
  AVG(position_change) AS avg_position_change,
  COUNT(*) AS race_count
FROM base
GROUP BY driver_name
HAVING race_count >= 5
ORDER BY avg_position_change DESC
LIMIT 10;
```

The results show that the drivers with the highest average position gains tend to achieve these improvements without relying on pit-stop strategies. Jean-Marc Gounon leads with an average position gain of +11.0 places across seven races, followed by David Brabham (+8.73), Éric Bernard (+7.08), and Nicolas Kiesa (+6.60). These drivers consistently moved forward during races despite starting from lower-grid positions.

A notable pattern in the data is that almost all drivers in the top-gain group recorded an average of 0.00 pit stops across their races. This indicates that their position improvements were not driven by strategic pit-stop timing or undercut attempts. Instead, their gains most likely resulted from race-pace advantages, retirements of other drivers, or simply starting at the very back of the grid and benefiting from low-pressure conditions. The only partial exception is Karun Chandhok, who averaged 0.27 pit stops per race and recorded an unusually long average pit-stop duration due to a single slow stop, yet his average position gain remained a fixed +5.0.

Based on the available data, the drivers with the largest average position gains did not rely on pit stops, as nearly all of them recorded zero pit-stop activity. Therefore, within this subset of races, position improvements appear unrelated to pit-stop frequency or duration.

### **4.2 How do pit-stop efficiency and reliability together influence championship performance**

**Objective:** The goal of this analysis is to see whether teams that are both reliable and fast in the pit lane score more championship points. We want to understand if having fewer mechanical problems and quicker pit stops together gives teams a real advantage over a full season.

**Method:** Using the master_clean table, we calculated each team’s average pit-stop time, reliability rate (how often they finished races), and average team points for every season. We compared these numbers across all teams to see which ones scored the most points. We then looked for patterns to find out whether teams with both fast pit stops and high reliability also ranked higher in the championship.

**Insight Goal:** The goal is to learn whether strong results come from a mix of fast pit stops and reliable cars, or if one factor matters more than the other. This helps explain whether championship success in Formula 1 depends mainly on strategy in the pits, having a dependable car, or a combination of both.

```python
%%bigquery --project=ba775-b09 --location=US
WITH team_metrics AS (
  SELECT
    year,
    constructor_name,
    ROUND(AVG(avg_pit_time), 2) AS avg_pit_time_ms,
    ROUND(SUM(CASE WHEN reliability_status = 'Finished' THEN 1 ELSE 0 END) / COUNT(*) * 100, 2) AS reliability_rate,
    ROUND(AVG(team_points), 2) AS avg_team_points
  FROM `ba775-b09.formula_one.master_clean`
  WHERE year >= 2000
  GROUP BY year, constructor_name
)
SELECT
  CORR(avg_pit_time_ms, avg_team_points) AS corr_pit_points,
  CORR(reliability_rate, avg_team_points) AS corr_reliability_points
FROM team_metrics;
```

The results show two important findings. First, the correlation between pit-stop time and team points is 0.309, which is considered weak. This means that faster pit stops only have a small effect on how many points a team earns in a season. Teams with quick pit stops may gain some advantage, but pit-stop speed is not a major factor in overall success.

Second, the correlation between reliability and team points is 0.774, which is very strong. This shows that teams who finish more races earn many more points on average. Reliability has a much larger impact on performance because finishing races gives teams more chances to score, while DNFs take away important opportunities.

Overall, the analysis shows that reliability matters much more than pit-stop speed when it comes to scoring championship points. Fast pit stops can help, but a reliable car is the key to consistent success throughout the season.
